```python
import pandas as pd
import seaborn as se

```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding='latin-1')
```

# converting date col

```python
data .info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data.info()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Year'] = data['Order Date'].dt.year
data['Order Month'] = data['Order Date'].dt.month
data['Order Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly sales

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
import matplotlib.pyplot as plt
```

```python
se.lineplot (x ='Order Month', y ='Sales',data= sales_by_month)
plt. title ('Monthly Sales Analysis')
plt.show()

```

# Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
categories = sales_by_category['Category'] 
sales = sales_by_category['Sales']         
plt.pie(sales, labels=categories, autopct='%1.1f%%')
plt.title('Sales Analysis by Category')
plt.show()
```

# sales by sub-category

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
se.barplot (x ='Sub-Category', y ='Sales',data= sales_by_subcategory)
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt. title ('Sales Analysis by Category')
plt.show()
```

# monthly profit

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
```

```python
profit_by_month
```

```python
se.lineplot (x ='Order Month', y ='Profit',data= profit_by_month )
plt.xticks(ticks=range(1, 13))
plt.tight_layout()
plt. title ('Monthly Profit Analysis')
plt.show()
```

# profit by  Category

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
```

```python
profit_by_category 
```

```python
categories = profit_by_category['Category'] 
profits = profit_by_category['Profit']       
plt.pie(profits, labels=categories, autopct='%1.1f%%')
plt.title('Profit Analysis by Category')
plt.show()
```

# profit by sub- category

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
se.barplot (x ='Sub-Category', y ='Profit',data= profit_by_subcategory)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt. title ('profit by Sub-Category')
plt.show()
```

# sales and profit- customer segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
plt.figure(figsize=(10, 6))

# Create a barplot for Sales by Segment
se.barplot(x='Segment', y='Sales', data=sales_profit_by_segment, color='skyblue', label='Sales')

# Add Profit barplot with a different color (overlay)
se.barplot(x='Segment', y='Profit', data=sales_profit_by_segment, color='lightgreen', label='Profit')

# Add a title and labels
plt.title('Sales and Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Amount')
plt.legend(title='Legend')

# Show the plot
plt.show()
```

# sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_profit_ ratio']= sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Sales','Profit']])
```
